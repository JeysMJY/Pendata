# Business Understanding

Fase Pemahaman Bisnis (Business Understanding) adalah fase awal yang krusial dalam siklus proyek data mining. Fase ini berfokus pada pemahaman proyek dan kebutuhan dari perspektif bisnis, yang kemudian diterjemahkan menjadi definisi masalah data mining secara teknis beserta rencana penyelesaiannya.


---

Berikut adalah catatan materi dari fase Pemahaman Bisnis:

## 1. Menentukan Tujuan Bisnis (Determine Business Objectives)

Tujuan utama pada langkah ini adalah memahami secara menyeluruh apa yang benar-benar ingin dicapai oleh pelanggan atau organisasi dari perspektif bisnis. Mengabaikan pemahaman ini dapat menyebabkan tenaga terbuang sia-sia karena memberikan jawaban yang benar untuk pertanyaan yang salah.

* **Latar Belakang (Background):** Mendokumentasikan situasi bisnis organisasi saat proyek dimulai. Ini mencakup identifikasi area masalah, struktur organisasi, peran orang-orang kunci, sasaran kelompok, dan solusi yang sudah ada saat ini.
* **Tujuan Bisnis (Business Objectives):** Mendeskripsikan sasaran utama bisnis (misalnya "mempertahankan pelanggan agar tidak pindah ke pesaing") beserta pertanyaan bisnis turunan lainnya (misalnya "apakah penurunan biaya administrasi memengaruhi retensi pelanggan bernilai tinggi?").
* **Kriteria Keberhasilan Bisnis (Business Success Criteria):** Mendefinisikan tolak ukur keberhasilan proyek dari kacamata bisnis. Kriteria ini bisa bersifat spesifik dan dapat diukur secara objektif (seperti mengurangi tingkat churn hingga persentase tertentu), atau bersifat subjektif (seperti "memberikan wawasan yang berguna").


## 2. Menilai Situasi dan Kondisi (Assess Situation)

Langkah ini adalah proses pencarian fakta yang lebih mendetail mengenai semua sumber daya, kendala, asumsi, dan faktor eksternal maupun internal yang perlu dipertimbangkan untuk proyek.

* **Inventaris Sumber Daya:** Mendata personel (ahli data, ahli bisnis), ketersediaan data, serta infrastruktur perangkat keras dan lunak.
* **Persyaratan, Asumsi, dan Kendala:** Mencatat tenggat waktu, masalah keamanan/hukum, serta asumsi-asumsi yang diambil selama proyek.
* **Risiko dan Kontinjensi:** Mengidentifikasi kemungkinan hambatan (seperti kualitas data yang buruk) dan menyiapkan rencana cadangan.
* **Analisis Biaya dan Manfaat:** Menilai apakah potensi keuntungan finansial atau strategis sebanding dengan biaya investasi proyek.

## 3. Menentukan Tujuan Data Mining (Determine Data Mining Goals)
Menerjemahkan tujuan bisnis ke dalam istilah teknis analisis data.

* **Tujuan Data Mining:** Menentukan output teknis yang dibutuhkan (misalnya: membangun model klasifikasi untuk memprediksi risiko kredit atau melakukan klastering untuk segmentasi pasar).
* **Kriteria Keberhasilan Data Mining:** Menentukan standar teknis performa model, seperti tingkat akurasi (accuracy), presisi (precision), atau tingkat kesalahan minimal yang dapat ditoleransi.

## 4. Membuat Rencana Proyek (Produce Project Plan)
Menyusun strategi dan langkah-langkah kerja untuk mencapai tujuan yang telah ditetapkan.

* **Rencana Proyek:** Menyusun jadwal kerja (roadmap), durasi tiap fase, ketergantungan antar tugas, dan titik evaluasi proyek.
* **Penilaian Awal Alat dan Teknik:** Memilih algoritma, teknik pemodelan, serta perangkat lunak yang akan digunakan untuk eksekusi proyek.
