# Data Understanding

**Informasi Dataset:**
- **Nama Dataset:** Pima Indians Diabetes Database
- **Jumlah Fitur:** 8 fitur (kolom prediktor)
- **Jumlah Kelas:** 2 kelas (0 = Negatif/Tidak Diabetes, 1 = Positif/Diabetes)
- **Jumlah Data:** 768 baris
- **Target / Label:** Kolom `Outcome`

**Deskripsi Kolom:**

| Nama Kolom | Tipe Data | Penjelasan Fitur |
|---|---|---|
| Pregnancies | Integer | Jumlah kehamilan yang pernah dialami pasien |
| Glucose | Integer | Konsentrasi glukosa plasma 2 jam dalam tes toleransi glukosa oral |
| BloodPressure | Integer | Tekanan darah diastolik (mm Hg) |
| SkinThickness | Integer | Ketebalan lipatan kulit trisep (mm) |
| Insulin | Integer | Insulin serum 2 jam (mu U/ml) |
| BMI | Float | Indeks massa tubuh (berat dalam kg / (tinggi dalam meter)^2) |
| DiabetesPedigreeFunction | Float | Fungsi silsilah diabetes (indikator riwayat genetik keluarga) |
| Age | Integer | Usia (tahun) |
| Outcome | Integer | Variabel kelas (Target/Label). 0 berarti tidak diabetes, 1 berarti diabetes. |
