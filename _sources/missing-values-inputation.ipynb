# Missing Values Inputation

Dalam dataset atau sekumpulan data, seringkali terdapat nilai yang hilang atau tidak lengkap. Sehingga perlu dilakukan imputasi untuk mengisi nilai yang hilang agar data dapat digunakan untuk analisis atau pemodelan. Berikut adalah beberapa metode imputasi yang umum digunakan:
1. Mean Imputation: Mengganti nilai yang hilang dengan rata-rata dari kolom tersebut.
2. Median Imputation: Mengganti nilai yang hilang dengan median dari kolom tersebut.
3. Mode Imputation: Mengganti nilai yang hilang dengan nilai yang paling sering muncul (modus) dalam kolom tersebut.
4. KNN Imputation: Mengganti nilai yang hilang dengan nilai yang paling sering muncul di antara k tetangga terdekat.
5. Regression Imputation: Menggunakan model regresi untuk memprediksi nilai yang hilang berdasarkan variabel lain.
6. WKNN Imputation: Mengganti nilai yang hilang dengan nilai yang paling sering muncul di antara k tetangga terdekat, tetapi memberikan bobot lebih pada tetangga yang lebih dekat.
7. Dan lain-lain.

Tetapi selain mengganti datanya, kita juga bisa menghapus data yang memiliki missing values, seperti beberapa metode berikut:
1. Menghapus baris yang memiliki missing values.
2. Menghapus kolom yang memiliki missing values.

Namun, ada beberapa syarat yang harus dipenuhi contohnya seperti berikut:
1. Jika kita menghapus baris, pastikan bahwa jumlah baris yang dihapus tidak terlalu banyak sehingga tidak merusak representasi data.
2. Jika kita menghapus kolom, pastikan bahwa kolom tersebut tidak terlalu penting contohnya kolom yang isinya laki-laki semuanya atau perempuan semuanya, sehingga tidak memberikan informasi yang berguna untuk analisis atau pemodelan.
3. Pastikan bahwa data yang tersisa setelah penghapusan masih cukup untuk analisis atau pemodelan yang diinginkan.

Pada materi ini kita akan membahas dan melakukan inputasi missing valus menggunakan WKNN Imutation.

# WKNN Imputation

Metode Weighted KNN memilih instans dengan nilai yang serupa terhadap instans yang tidak lengkap, tetapi memberikan bobot lebih pada tetangga yang lebih dekat. Ini dapat menghasilkan imputasi yang lebih akurat dibandingkan dengan KNN biasa, terutama jika data memiliki pola yang kompleks. 
Ukuran kesamaan $s_i (y_i)$ antara dua contoh $y_i$ dan $y_j$ didefinisikan olej jarak Euclidian yang dihitung dari atribut-atribut yang terdapat pada $y_i$.

Berikut mendefinisikan ukuran tersebut sebagai berikut:

$$1/s_i = \sum_{h_i \in O_i \cap O_j} (y_{ih} - y_{jh})^2$$

Rumus ini adalah cara untuk mengukur seberapa mirip dua data (misalnya Data A dan Data B) dengan aturan berikut:
- Menghitung Selisih (Jarak Euclidean): Kita melihat perbedaan antara ciri-ciri yang ada di Data A ($y_i$) dengan Data B ($y_j$). Selisihnya dikuadratkan agar hasilnya selalu positif dan menekankan perbedaan yang besar.

- Hubungan Terbalik ($1/s_i$): Jarak dan Kemiripan itu berbanding terbalik.
  - Jika jaraknya kecil (selisihnya sedikit), maka kemiripannya ($s$) besar.
  - Jika jaraknya besar (selisihnya banyak), maka kemiripannya ($s$) kecil.

- Hanya Bandingkan yang Ada ($O_i \cap O_j$): Kita hanya menghitung bagian yang sama-sama memiliki data. Jika salah satu data kosong (NaN), bagian itu diabaikan agar tidak merusak perhitungan.

Kesimpulannya:

Semakin kecil total selisihnya, maka kedua data tersebut dianggap semakin kembar atau identik.

Entri data yang hilang ($y_{ih}$) diimputasi dengan menggunakan rata-rata tertimbang dari nilai-nilai yang sesuai dari tetangga-tetangga terdekat, dengan bobot yang diberikan oleh ukuran kesamaan $s_i$.

$$\hat{y}_{ih} = \frac{\sum_{j \in I_{Kih}} s_i(y_j)y_{jh}}{\sum_{j \in I_{Kih}} s_i(y_j)}$$



di mana $I_{Kih}$ adalah himpunan indeks dari tetangga terdekat (KNN) untuk data ke-$i$. Jika nilai $y_{jh}$ ternyata juga hilang, maka data tersebut tidak disertakan dalam perhitungan $I_{Kih}$. Perlu diingat bahwa metode KNNI tidak punya rumus pasti untuk menentukan jumlah tetangga ($K$) yang paling bagus; nilai $K$ harus dicari melalui percobaan langsung (secara empiris).

Ini adalah Rata-rata Tertimbang (Weighted Average).
- Pembilang (Atas): Menjumlahkan semua hasil kali antara Bobot Kemiripan ($s_i$) dengan Nilai dari Tetangga ($y_{jh}$).
- Penyebut (Bawah): Menjumlahkan seluruh Bobot Kemiripan ($s_i$) yang digunakan sebagai angka pembagi (normalisasi).
- Logikanya: Jika seorang tetangga punya kemiripan yang tinggi (bobot $s_i$ besar), maka angka milik tetangga tersebut ($y_{jh}$) akan sangat menentukan hasil tebakan akhir. Sebaliknya, jika tetangga kurang mirip (bobot kecil), pengaruhnya pun akan kecil.

Penentuan jumlah tetangga ($K$) yang optimal sangat penting untuk mendapatkan hasil imputasi yang baik. Nilai $K$ yang terlalu kecil dapat membuat imputasi menjadi terlalu sensitif terhadap noise, sementara nilai $K$ yang terlalu besar dapat menyebabkan imputasi menjadi terlalu umum dan kurang akurat. Oleh karena itu, pemilihan $K$ biasanya dilakukan melalui percobaan langsung (empiris) dengan menggunakan teknik validasi silang (cross-validation) untuk mengevaluasi kinerja model pada data yang tidak terlihat.

Langkah-langkah untuk melakukan WKNN Imputation:
1. Pilih nilai $K$ yang sesuai untuk jumlah tetangga terdekat.
2. Hitung jarak dan kemiripan $(s_i)$ antara data yang memiliki nilai yang hilang dengan data lainnya.
3. Pilih $K$ tetangga terdekat berdasarkan kemiripan yang dihitung.
4. Hitung Estimasi nilai yang hilang menggunakan weighted average dari nilai-nilai tetangga yang dipilih.
5. Ulangi proses ini untuk setiap nilai yang hilang dalam dataset.

Selanjutnya kita akan membahas normalisasi data sebelum melakukan imptasi WKNN.

# Data Normalization

Data Normalization adalah proses mengubah skala data sehingga semua fitur memiliki rentang nilai yang sama. Ini penting untuk metode seperti KNN atau WKNN karena menggunakan jarak untuk menentukan kemiripan antara data, dan jika fitur memiliki skala yang berbeda, maka fitur dengan skala yang lebih besar dapat mendominasi perhitungan jarak.

Beberapa metode normalisasi yang umum digunakan adalah:

## Min-Max Normalization
Metode ini mengubah nilai fitur ke dalam rentang [0, 1] dengan menggunakan rumus berikut:

$$x' = \frac{x - min(x)}{max(x) - min(x)}$$

di mana:
- $x'$ = nilai yang telah dinormalisasi.
- $x$ = nilai asli dari fitur.
- $min(x)$ = nilai minimum dari fitur.
- $max(x)$ = nilai maksimum dari fitur.

Contoh:

Data awal:
| ID | NIlai ($x$) |
|----|-------------|
| 1  | 20          |
| 2  | 50          |
| 3  | 80          |
| 4  | 120         |
| 5  | 150         |

Proses Normalisasi:

Diketahui $min(x) = 20$ dan $max(x) = 150$.

Rumus:

$$x' = \frac{x - min(x)}{max(x) - min(x)}$$

Perhitungan (ID 2):

$$x' = \frac{50 - 20}{150 - 20} = \frac{30}{130} = \mathbf{0,2308}$$

Setelah normalisasi:
| ID | NIlai ($x$) |
|----|-------------|
| 1  | 0          |
| 2  | 0,2308          |
| 3  | 0,4615          |
| 4  | 0,7692      |
| 5  | 1          |

## Z-Score Normalization

Metode ini digunakan jika data memiliki sebaran yang sangat beragam atau mengandung pencilan (outliers), dengan rumus sebagai berikut:

$$x' = \frac{x - \mu}{\sigma}$$

di mana:
- $x'$ = nilai yang telah dinormalisasi.
- $x$ = nilai asli dari fitur.
- $\mu$ = rata-rata dari fitur.
- $\sigma$ = standar deviasi dari fitur.

Contoh:
Data awal:
| ID | NIlai ($x$) |
|----|-------------|
| 1  | 20          |
| 2  | 50          |
| 3  | 80          |
| 4  | 120         |
| 5  | 150         |

Proses Normalisasi:
1. Mencari rata-rata($\mu$)
Rata-rata dihitung dengan menjumlahkan semua nilai, lalu dibagi jumlah data ($n=5$).

$$\mu = \frac{20 + 50 + 80 + 120 + 150}{5} = \frac{420}{5} = 84$$

2. Mencari standar deviasi ($\sigma$)
Standar deviasi dihitung dengan rumus:

$$\sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2}{n}}$$

Perhitungan:

$$\sigma = \sqrt{\frac{(20-84)^2 + (50-84)^2 + (80-84)^2 + (120-84)^2 + (150-84)^2}{5}}$$

$$\sigma = \sqrt{\frac{4096 + 1156 + 16 + 1296 + 4356}{5}}$$

$$\sigma = \sqrt{\frac{10920}{5}} = \sqrt{2184} = 46.73$$

3. Normalisasi (ID 2):

$$x' = \frac{x - \mu}{\sigma}$$

$$x' = \frac{50 - 84}{46.73} = \frac{-34}{46.73} = \mathbf{-0,7275}$$

Setelah normalisasi:
| ID | NIlai ($x$) |
|----|-------------|
| 1  | -1,3695          |
| 2  | -0,7275        |
| 3  | -0,0856          |
| 4  | 0,7703        |
| 5  | 1,4123         |


## Decimal Scaling Normalization

Metode ini mengubah nilai fitur dengan membagi nilai asli dengan faktor 10 yang dipangkatkan dengan jumlah digit maksimum dari nilai dalam fitur tersebut, dengan rumus sebagai berikut:

$$x' = \frac{x}{10^j}$$

di mana:

- $x'$ = nilai yang telah dinormalisasi.
- $x$ = nilai asli dari fitur.
- $j$ = bilangan bulat terkecil sedemikian hingga $max(|x'|) < 1$.

Contoh:

Data awal:
| ID | NIlai ($x$) |
|----|-------------|
| 1  | 20          |
| 2  | 50          |
| 3  | 80          |
| 4  | 120         |
| 5  | 150         |

Proses Normalisasi:

1. Cari nilai $j$:
Tentukan angka terbesar untuk tahu berapa kali koma harus digeser ke kiri agar hasilnya di bawah 1.
- Data: 20, 50, 80, 120, 150.
- Karena 150 punya 3 digit, maka kita butuh $j=3$ ($10^3 = 1000$).
- $150 / 1000 = 0,15$ (Sudah di bawah 1).

2. Normalisasi (ID 2):

$$x' = \frac{x}{10^j}$$

$$x' = \frac{50}{10^3} = \frac{50}{1000} = \mathbf{0,05}$$

Setelah normalisasi:
| ID | Nilai ($x$) |
|----|-------------|
| 1  | 0,02        |
| 2  | 0,05        |
| 3  | 0,08        |
| 4  | 0,12        |
| 5  | 0,15        |




```python
# Implementasi normalisasi pada python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Menyiapkan Data Awal
data = {
    'ID': [1, 2, 3, 4, 5],
    'Nilai': [20, 50, 80, 120, 150]
}
df = pd.DataFrame(data)

# Sklearn membutuhkan input dalam bentuk array 2D (reshape)
nilai_array = df['Nilai'].values.reshape(-1, 1)

# 2. Min-Max Normalization
min_max_scaler = MinMaxScaler()
df['Min-Max'] = min_max_scaler.fit_transform(nilai_array)

# 3. Z-Score Normalization
z_score_scaler = StandardScaler()
df['Z-Score'] = z_score_scaler.fit_transform(nilai_array)

# 4. Decimal Scaling (Manual/Numpy karena tidak ada di sklearn)
max_abs = np.max(np.abs(nilai_array))
j = np.ceil(np.log10(max_abs))
df['Decimal'] = df['Nilai'] / (10**j)

# Menampilkan Hasil
print("Hasil Normalisasi dengan Library:")
print(df.to_string(index=False))
```

## Langkah 2: Hitung Jarak dan Kemiripan ($s_i$)

Pada langkah ini, saya akan menggunakan rumus WKNN seperti yang sudah dijelaskan sebelumnya untuk menghitung jarak dan kemiripan antara data yang memiliki nilai yang hilang dengan data lainnya. Saya akan menggunakan rumus berikut:

$$1/s_i = \sum_{h_i \in O_i \cap O_j} (y_{ih} - y_{jh})^2$$

Karena data yang hilang berada pada baris ke-7, maka saya buat:
- $y_ih$ = nilai pada kolom JML untuk data ke-i (Data 1 - 6)
- $y_jh$ = nilai pada kolom JML untuk data ke-j (Data 7)

Kolom yang dhitung adalah kolom IPK dan PO karena kolom JML yang akan diimputasi.

Berikut adalah untuk semua data:

|No | Selisih $\sum (y_{ih} - y_{jh})^2$ | Kemiripan ($s_i = 1/\sum (y_{ih} - y_{jh})^2$) |
|---|-------------------------|--------------------|
| 1 | $(0 - 0)^2 + (0 - 0,5)^2$ = 0,25 | 1/0,25 = 4 |
| 2 | $(0,5 - 0)^2 + (0,5 - 0,5)^2$ = 0,25 | 1/0,25 = 4 |
| 3 | $(1 - 0)^2 + (0 - 0,5)^2$ = 1,25 | 1/1,25 = 0,8 |
| 4 | $(0 - 0)^2 + (0 - 0,5)^2$ = 0,25 | 1/0,25 = 4 |
| 5 | $(0,5 - 0)^2 + (0,5 - 0,5)^2$ = 0,25 | 1/0,25 = 4 |
| 6 | $(1 - 0)^2 + (1 - 0,5)^2$ = 1,25 | 1/1,25 = 0,8 |

```python
# 1. Memisahkan data referensi (baris 1-6) dan data target (baris 7)
referensi = df.iloc[:6].copy()
target = df.iloc[6]

# 2. Menghitung selisih kuadrat (IPK dan PO)
referensi["sum_sq"] = (referensi["IPK"] - target["IPK"])**2 + (referensi["PO"] - target["PO"])**2

# 3. Menghitung Kemiripan (si)
referensi["si"] = 1 / referensi["sum_sq"]

print("hasil Menghitung Jarak dan Kemiripan")
print(referensi[['NO', 'sum_sq', 'si']].to_string(index=False))
```

## Langkah 3: Hitung Estimasi Menggunakan Weighted Average


Dalam menghitung estimasi kita menggunakan rumus Weighted Average yang telah dijelaskan sebelumnya:

$$\hat{y}_{ih} = \frac{\sum_{j \in I_{Kih}} s_i(y_j)y_{jh}}{\sum_{j \in I_{Kih}} s_i(y_j)}$$

Dengan mengalikan kemiripan ($s_i$) dengan nilai JML dari data yang memiliki nilai JML (Data 1 - 6) dan kemudian menjumlahkannya, lalu dibagi dengan jumlah kemiripan untuk mendapatkan estimasi nilai JML yang hilang pada Data 7.


### Pembilang (Atas):

- Data 1: $4 \times 0 = 0$
- Data 2: $4 \times 1 = 4$
- Data 3 : $0,8 \times 0 = 0$
- Data 4 : $4 \times 1 = 4$
- Data 5 : $4 \times 0 = 0$
- Data 6 : $0,8 \times 1 = 0,8$

Total Pembilang = $0 + 4 + 0 + 4 + 0 + 0,8 = 8,8$

### Penyebut (Bawah):
- $4 + 4 + 0,8 + 4 + 4 + 0,8 = 17,6$

### Hasil Akhir:
$$\hat{y}_{ih} = \frac{8,8}{17,6} = 0,5$$

Berikut adalah data setelah dilakukan imputasi:

| NO | IPK | PO | JML|
|----|-----|----|----|
| 1  | 0   | 0  | 0  |
| 2  | 0,5 | 0,5| 1  |      
| 3  | 1 | 0| 0  |      
| 4  | 0 | 0| 1  |      
| 5  | 0,5 | 0,5| 0  |      
| 6  | 1 | 1| 1  |      
| 7  | 0 | 0,5| **0,5**  |

```python
# Implementasi Estimasi Nilai Menggunakan Weighted Average

# 1. Menghitung Pembilang
sum_si_yi = (referensi["si"] * referensi["JML"]).sum()

# 2. Menghitung Penyebut
sum_si = referensi["si"].sum()

# 3. Menghitung Nilai Estimas
estimasi_jml = sum_si_yi / sum_si

# 4. Memasukkan hasil
df.loc[6, "JML"] = estimasi_jml

print("Hasil menghitung estimasi nilai")
print(f"Total Pembilang : {sum_si_yi:.1f}")
print(f"Total Penyebut: {sum_si:.1f}")
print(f"Hasil Akhir: {estimasi_jml}")
print()
print("Hasil Akhir Data setelah imputasi:")
print(df)
```

Berdasarkan perhitungan di atas, estimasi nilai JML yang hilang pada Data 7 adalah **0,5** setelah dilakukan imputasi menggunakan metode WKNN dengan normalisasi data menggunakan min-max normalization.
